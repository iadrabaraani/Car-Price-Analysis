```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv("car_prices.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

```python
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(exclude=np.number).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df = df.drop_duplicates()
```

# Data Cleaning Summary
Missing values were handled using median for numerical columns and mode for categorical columns. 
Duplicate records were removed to ensure data accuracy and consistency for analysis.

# Key Observations from Data Queries
. Car prices vary significantly across brands and models.
. Newer cars and cars with better condition tend to have higher selling prices.
. Mileage (odometer reading) shows a negative relationship with selling price.
. Certain states consistently show higher average selling prices for newer cars.


```python
plt.figure(figsize=(10,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

# Insight:  
Selling price shows a strong negative correlation with odometer and a positive correlation with year and condition indicating that newer and better maintained cars are priced higher.


```python
df.groupby('year')['sellingprice'].mean().plot(marker='o')
plt.title("Average Selling Price by Year")
plt.show()
```

# Insight:
The average selling price increases with newer model years which reflects higher demand and value for recent vehicles.


```python
df.groupby('odometer')['sellingprice'].mean().plot()
plt.title("Selling Price vs Odometer")
plt.show()
```

# Insight:
As odometer readings increases the average selling price decreases confirming depreciation with higher vehicle usage.

```python
df['state'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Cars Sold per State")
plt.show()
```

# Insight:
A few states dominated the number of car sales indicating stronger used car markets and higher transaction volumes in these regions.

```python
df['condition_bin_5'] = pd.cut(df['condition'], bins=range(0,101,5))
df.groupby('condition_bin_5', observed=False)['sellingprice'].mean().plot(kind='bar')
plt.show()
```

# Insight:  
Cars with higher condition scores command significantly higher prices showing that vehicle condition is a key pricing factor.


```python
df['condition_bin_10'] = pd.cut(df['condition'], bins=range(0,101,10))
df['condition_bin_10'].value_counts().sort_index().plot(kind='bar')
plt.show()
```

# Insight:  
Most cars fall within mid range condition scores suggesting a balanced distribution between heavily used and well maintained vehicles.


```python
plt.figure(figsize=(12,5))
sns.boxplot(x='color', y='sellingprice', data=df)
plt.xticks(rotation=45)
plt.show()
```

# Insight: 
Selling prices vary across different car colors with some colors showing higher median prices and greater price variability.
